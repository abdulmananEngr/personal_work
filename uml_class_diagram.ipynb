```python
import spacy
from spacy import displacy
from graphviz import Digraph
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

```python
# Function to extract noun phrases from the text
def extract_noun_phrases(text):
    doc = nlp(text)
    noun_phrases = []
    for chunk in doc.noun_chunks:
        noun_phrases.append(chunk.text)
    return noun_phrases

# Function to generate UML diagram from noun phrases
def generate_uml_diagram(noun_phrases):
    dot = Digraph(comment='UML Diagram')
    for noun_phrase in noun_phrases:
        dot.node(noun_phrase)
    return dot

# Function to generate class diagram from noun phrases
def generate_class_diagram(noun_phrases):
    dot = Digraph(comment='Class Diagram')
    for noun_phrase in noun_phrases:
        dot.node(noun_phrase, shape='box')
    return dot
```

```python
# Text input
text = "The car has a color, an engine, and four wheels. The engine has a power source."
```

```python
# Extract noun phrases from the text
noun_phrases = extract_noun_phrases(text)
```

```python
noun_phrases
```

```python
# Generate UML diagram
uml_diagram = generate_uml_diagram(noun_phrases)
```

```python
uml_diagram
```

```python
# Save and render the UML diagram
uml_diagram.render('uml_diagram', format='png', view=True)
```

```python
# Generate class diagram
class_diagram = generate_class_diagram(noun_phrases)
```
